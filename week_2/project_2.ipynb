```python
import numpy as np

def polynomial_solver():
    print("Which equation would you like to solve?")
    print("1. Quadratic (Ax^2 + Bx + C = 0)")
    print("2. Cubic (Ax^3 + Bx^2 + Cx + D = 0)")
    print("3. Quartic (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")
    
    choice = input("\nEnter your choice (1, 2, or 3): ")
    
    # Determine the required variables based on the user's choice
    if choice == '1':
        variables = ['A', 'B', 'C']
    elif choice == '2':
        variables = ['A', 'B', 'C', 'D']
    elif choice == '3':
        variables = ['A', 'B', 'C', 'D', 'E']
    else:
        print("Invalid choice. Mission aborted.")
        return

    # Collect coefficients from the user
    print("\nEnter the coefficients:")
    coefficients = []
    for var in variables:
        val = float(input(f"{var} = "))
        coefficients.append(val)
        
    # Calculate roots using numpy's built-in polynomial root finder
    roots = np.roots(coefficients)
    
    # Display the results
    print("\n--- Results ---")
    print("The roots are:")
    for i, root in enumerate(roots, 1):
        # Using np.round to keep the output clean
        print(f"Root {i}: {np.round(root, 4)}")

# Run the program
if __name__ == "__main__":
    polynomial_solver()
```
